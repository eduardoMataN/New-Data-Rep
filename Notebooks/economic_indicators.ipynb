```python
import pandas as pd
import plotly.express as px
import dash as dash
from dash import dcc
from dash import html
import dash_bootstrap_components as dbc
from dash.dependencies import Input, Output, State

```

```python
from ctypes import alignment


app= dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
df=pd.read_excel('Test Dataset.xlsx')
df_copy=df.copy()
initialValue=df_copy.loc[df_copy.County=='Hidalgo County', 'Current'].tolist()
initialValue2=df_copy.loc[df_copy.County=='Hidalgo County', 'Current'].tolist()



app.layout=html.Div(children=[
    dbc.NavbarSimple(children=[
    dbc.NavItem(dbc.NavLink('Demographic Indicators', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Economic Indicators', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Border Security', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Education', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Social Indicators', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Transportation', href='#', style={'color':'white'}))
        ],
        brand='HIBRED',
        brand_href='#',
        color='#041E42',
        style={'font-weight':'bold'},
        className='navbar-dark'
        ),
    dbc.Container(children=[
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.H1(id='County-Title', children='Hidalgo County', style={'color':'#041E42'})
                    
                ]),
            ),
            dbc.Col(
                html.Div([
                    html.H1(id='County-Title2', children='Hidalgo County', style={'color':'#041E42'})
                ])
            )
        ])
    ]),
    html.Br(),
    dbc.Container(children=[
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.H3(id='Number1', children=initialValue[0], style={'color':'#FF8200'})
                ])
            ),
            dbc.Col(
                html.Div([
                    html.H3(id='Number2', children=initialValue2[0], style={'color':'#FF8200'})
                ])
            )
        ]),
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.H3(id='Indic-Title', children='Current Total Families', style={'color':'#FF8200'})
                ])
            ),
            dbc.Col(
                html.Div([
                    html.H3(id='Indic-Title2', children='Current Total Families', style={'color':'#FF8200'})
                ])
            )
        ])
    ]),
    html.Br(),
    dbc.Container(children=[
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.Label(['Indicator'], style={'font-weight':'bold'}),
                    dcc.Dropdown(id='select-indicator',
                                options=[{'label':x, 'value':x} for x in sorted(df.Indicator.unique())],
                                multi=False,
                                value='Total Families',
                                style={'width':'80%'})
                ])
            ),
            
            dbc.Col(
                html.Div([
                    html.Label(['Indicator'], style={'font-weight':'bold'}),
                    dcc.Dropdown(id='select-indicator2',
                        options=[{'label':x, 'value':x} for x in sorted(df.Indicator.unique())],
                        multi=False,
                        value='Total Families',
                        style={'width':'80%'})
                ])
            ),
            
        ]),
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.Label(['County'], style={'font-weight':'bold'}),
                    dcc.Dropdown(id='select-county',
                        options=[{'label':i, 'value':i} for i in sorted(df.County.unique())],
                        multi=True,
                        value=[x for x in sorted(df.County.unique())],
                        style={'width':'100%', 'height':'40%'}
                        )
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Label(['County'], style={'font-weight':'bold'}),
                    dcc.Dropdown(id='select-county2',
                        options=[{'label':x, 'value':x} for x in sorted(df.County.unique())],
                        multi=True,
                        value=[x for x in sorted(df.County.unique())],
                        style={'width':'100%', 'height':'40%'})
                ])
            )
        ], style={'height':'30%'})
    ]),
    dbc.Container(children=[
        dbc.Row([
            
            dbc.Col([
                html.Div([
                    dcc.Graph(id='graph', figure={})
                ]),
                html.Div([
                    dcc.RangeSlider(df['Year'].min(), df['Year'].max(), step=1,value=[df['Year'].min(), df['Year'].max()], id='slider1')
                ])
                ]),
            dbc.Col([
                html.Div([
                    dcc.Graph(id='graph2', figure={})
                ]),
                html.Div([
                    dcc.RangeSlider(df['Year'].min(), df['Year'].max(), 1, value=[df['Year'].min(), df['Year'].max()], id='slider2')
                ])
        ]),
            
        ])
    ])
    
    
    
    
    
    
]
)

@app.callback(
    [Output(component_id='graph', component_property='figure'), Output(component_id='graph2', component_property='figure'), 
    Output(component_id='County-Title', component_property='children'), Output(component_id='County-Title2', component_property='children'),
    Output(component_id='Number1', component_property='children'), Output(component_id='Number2', component_property='children'),
    Output(component_id='Indic-Title', component_property='children'), Output(component_id='Indic-Title2', component_property='children')],
    [Input(component_id='select-indicator', component_property='value'),
    Input(component_id='select-county', component_property='value'),
    Input(component_id='select-indicator2', component_property='value'),
    Input(component_id='select-county2', component_property='value')]
)
def update_indicator(indicator, county, indicator2, county2):
    dff=df.copy()
    dff=dff[dff['Indicator']==indicator]
    dff=dff[dff['County'].isin(county)]
    fig=px.line(dff, x='Year', y='Value', title=indicator+' by County ', color='County')
    fig.update_xaxes(nticks=len(pd.unique(dff['Year'])))
    dff2=df.copy()
    dff2=dff2[dff2['Indicator']==indicator2]
    dff2=dff2[dff2['County'].isin(county2)]
    fig2=px.line(dff2,x='Year', y='Value', title=indicator2+' by County ', color='County')
    fig2.update_xaxes(nticks=len(pd.unique(dff2['Year'])))
    dff3=df.copy()
    dff=dff[dff['Year']==dff['Year'].max()]
    dff2=dff2[dff2['Year']==dff2['Year'].max()]


    
    return fig, fig2, indicator, indicator2, dff['Value'].sum(), dff2['Value'].sum(), 'Current '+indicator, 'Current '+indicator2


```

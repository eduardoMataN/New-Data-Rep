```python
from ctypes import alignment

from matplotlib import animation


app= dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])
df=pd.read_excel('Test Dataset.xlsx')
df_copy=df.copy()
initialValue=df_copy.loc[df_copy.County=='Hidalgo County', 'Current'].tolist()
initialValue2=df_copy.loc[df_copy.County=='Hidalgo County', 'Current'].tolist()
df_copy=df[df['Indicator']=='Total Families']



app.layout=html.Div(children=[
    dbc.NavbarSimple(children=[
    dbc.NavItem(dbc.NavLink('Demographic Indicators', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Economic Indicators', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Border Security', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Education', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Social Indicators', href='#', style={'color':'white'})),
    dbc.NavItem(dbc.NavLink('Transportation', href='#', style={'color':'white'}))
        ],
        brand='HIBRED',
        brand_href='#',
        color='#041E42',
        style={'font-weight':'bold'},
        className='navbar-dark'
        ),
    dbc.Container(children=[
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.H1(id='County-Title', children='Hidalgo County', style={'color':'#041E42'})
                    
                ]),
            ),
            dbc.Col(
                html.Div([
                    html.H1(id='County-Title2', children='Hidalgo County', style={'color':'#041E42'})
                ])
            )
        ])
    ]),
    html.Br(),
    dbc.Container(children=[
        dbc.Row([
            dbc.Col(
                html.Div([
                    daq.LEDDisplay(id='Number1', value=initialValue[0], color='#FF8200')
                ])
            ),
            dbc.Col(
                html.Div([
                    daq.LEDDisplay(id='Number2', value=initialValue[0], color='#FF8200')
                ])
            )
        ]),
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.H3(id='Indic-Title', children='Current Total Families', style={'color':'#FF8200'})
                ])
            ),
            dbc.Col(
                html.Div([
                    html.H3(id='Indic-Title2', children='Current Total Families', style={'color':'#FF8200'})
                ])
            )
        ])
    ]),
    html.Br(),
    dbc.Container(children=[
        dbc.Row([
            dbc.Col(
                html.Div([
                    html.Label(['Indicator'], style={'font-weight':'bold'}),
                    dcc.Dropdown(id='select-indicator',
                                options=[{'label':x, 'value':x} for x in sorted(df.Indicator.unique())],
                                multi=False,
                                value='Total Families',
                                style={'width':'100%'})
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Label(['Initial Value'], style={'font-weight':'bold'}),
                      daq.NumericInput(
                        id='Y-Axes1-Start', min=df_copy['Value'].min(), max=df_copy['Value'].max(), value=df_copy['Value'].min(), size=80
                      )
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Label(['End Value'], style={'font-weight':'bold'}),
                      daq.NumericInput(
                        id='Y-Axes1-End', min=df_copy['Value'].min(), max=df_copy['Value'].max(), value=df_copy['Value'].max(), size=80
                      )
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Button('Reset', id='reset-button', n_clicks=0)
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Button('Reset', id='reset-button2', n_clicks=0)
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Label(['Indicator'], style={'font-weight':'bold'}),
                    dcc.Dropdown(id='select-indicator2',
                        options=[{'label':x, 'value':x} for x in sorted(df.Indicator.unique())],
                        multi=False,
                        value='Total Families',
                        style={'width':'100%'})
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Label(['Initial Value'], style={'font-weight':'bold'}),
                      daq.NumericInput(
                        id='Y-Axes2-Start', min=df_copy['Value'].min(), max=df_copy['Value'].max(), value=df_copy['Value'].min(), size=80
                      )
                ])
            ),
            dbc.Col(
                html.Div([
                    html.Label(['End Value'], style={'font-weight':'bold'}),
                      daq.NumericInput(
                        id='Y-Axes2-End', min=df_copy['Value'].min(), max=df_copy['Value'].max(), value=df_copy['Value'].max(), size=80
                      )
                ])
            )
            
        ]),

    ]),
    dbc.Container(children=[
        dbc.Row([
            
            dbc.Col([
                html.Div([
                    dcc.Graph(id='graph', figure={}, style={'height':'65vh', 'width':'60vh'})
                ]),
                
                ]),
            dbc.Col([
                html.Div([
                    dcc.Graph(id='graph2', figure={}, style={'height':'65vh', 'width':'60vh'})
                ]),
                
        ]),
            
        ])
    ])
    
    
    
    
    
    
]
)


@app.callback(
    [Output(component_id='graph', component_property='figure'), Output(component_id='graph2', component_property='figure'), 
    Output(component_id='County-Title', component_property='children'), Output(component_id='County-Title2', component_property='children'),
    Output(component_id='Number1', component_property='value'), Output(component_id='Number2', component_property='value'),
    Output(component_id='Indic-Title', component_property='children'), Output(component_id='Indic-Title2', component_property='children'),
    Output(component_id='Y-Axes1-Start', component_property='max'), Output(component_id='Y-Axes1-End', component_property='min'),
    Output(component_id='Y-Axes2-Start', component_property='max'), Output(component_id='Y-Axes2-End', component_property='min'),
    Output(component_id='reset-button', component_property='n_clicks'),
    Output(component_id='Y-Axes1-Start', component_property='value'), Output(component_id='Y-Axes1-End', component_property='value'),
    Output(component_id='Y-Axes2-Start', component_property='value'), Output(component_id='Y-Axes2-End', component_property='value'),
    Output(component_id='reset-button2', component_property='n_clicks')],
    [Input(component_id='select-indicator', component_property='value'),
    Input(component_id='select-indicator2', component_property='value'),
    Input(component_id='Y-Axes1-Start', component_property='value'), Input(component_id='Y-Axes1-End', component_property='value'),
    Input(component_id='Y-Axes2-Start', component_property='value'), Input(component_id='Y-Axes2-End', component_property='value'),
    Input(component_id='reset-button', component_property='n_clicks'), Input(component_id='reset-button2', component_property='n_clicks')]
)
def update_indicator(indicator,  indicator2, start1, end1, start2, end2, resetB, resetB2):
    dff=df.copy()
    dff=dff[dff['Indicator']==indicator]
    trigger_id=ctx.triggered_id
    if(trigger_id=='select-indicator'):
        resetB+=1
    if(resetB>0):
        start1=dff['Value'].min()
        end1=dff['Value'].max()
        resetB=0

    dff=dff[(dff['Value']>=start1) & (dff['Value']<=end1)]
    
    fig=px.line(dff, x='Year', y='Value', title=indicator+' by County ', color='County',  width=700, height=650)
    fig.update_xaxes(nticks=len(pd.unique(dff['Year'])), rangeslider_visible=True)
    


    dff2=df.copy()
    dff2=dff2[dff2['Indicator']==indicator2]
    if(trigger_id=='select-indicator2'):
        resetB2+=1
    if(resetB2>0):
        start2=dff2['Value'].min()
        end2=dff2['Value'].max()
        resetB2=0
    
    dff2=dff2[(dff2['Value']>=start2) & (dff2['Value']<=end2)]
    fig2=px.line(dff2,x='Year', y='Value', title=indicator2+' by County ', color='County',  width=700, height=650)
    fig2.update_xaxes(nticks=len(pd.unique(dff2['Year'])), rangeslider_visible=True)
    dff3=df.copy()
    dff=dff[dff['Year']==dff['Year'].max()]
    dff2=dff2[dff2['Year']==dff2['Year'].max()]


    
    return fig, fig2, indicator, indicator2, dff['Value'].sum(), dff2['Value'].sum(), 'Current '+indicator, 'Current '+indicator2, end1-1, start1+1, end2-1, start2+1, resetB, start1, end1, start2, end2, resetB2


```
